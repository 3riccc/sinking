```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成一个线条
# t_len总长度
# s_len单个线条的长度
# sigma纵向方差
# ybase基准的y
def gene_line(tlen,slen,sigma,ybase):
    # 开始位置
    slen = np.random.random()*slen
    x0 = np.random.random()*tlen
    x1 = x0+slen
    # 纵坐标
    y = np.random.normal(0, sigma)+ybase
    return [[x0-slen/2,x1-slen/2],[y,y]]
```

```python
poss = []
for i in range(2500):
    poss.append(gene_line(20,3,1.5,0))
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(16,10))
for pos in poss:
    plt.plot(pos[0],pos[1],color='black',alpha=0.8)
plt.xlim(0,20)
plt.ylim(-10,10)

plt.xticks(())
plt.yticks(())

# 取消边框
for key, spine in ax.spines.items():
    # 'left', 'right', 'bottom', 'top'
    if key == 'right' or key == 'top' or key == 'bottom' or key == 'left':
        spine.set_visible(False)
plt.show()
```

```python
possup = []
for i in range(1023):
    possup.append(gene_line(20,1.5,0.5,10))
possdown = []
for i in range(1023):
    possdown.append(gene_line(20,1.5,0.5,0))
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(16,10))
for pos in possup:
    plt.plot(pos[0],pos[1],color='black',alpha=0.5)
for pos in possdown:
    plt.plot(pos[0],pos[1],color='black',alpha=0.5)
plt.xlim(0,20)
plt.ylim(-5,15)

plt.xticks(())
plt.yticks(())

# 取消边框
for key, spine in ax.spines.items():
    # 'left', 'right', 'bottom', 'top'
    if key == 'right' or key == 'top' or key == 'bottom' or key == 'left':
        spine.set_visible(False)
plt.show()
```

```python
# 画一个圆
def get_circle(center,radius):
    thetas = np.arange(0,2*np.pi,0.01)
    xs = np.cos(thetas) * radius + center[0]
    ys = np.sin(thetas) * radius + center[1]
    return xs,ys
```

```python
xs,ys = get_circle([2,2],1)
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(5,5))
plt.plot(xs,ys)

plt.xticks(())
plt.yticks(())
plt.xlim(0,4)
plt.ylim(0,4)

# 取消边框
for key, spine in ax.spines.items():
    # 'left', 'right', 'bottom', 'top'
    if key == 'right' or key == 'top' or key == 'bottom' or key == 'left':
        spine.set_visible(False)
plt.show()
```
